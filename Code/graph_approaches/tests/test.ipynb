```
import dgl
import torch

from graph_transformer.display_graph import displayGraph
from graph_transformer.graph_dataset import loadGraphsAndSplit
from preprocessing.preprocessor import hashGraphContent
```

```
unique_graph_hashes = set()
def removeIdenticalGraphsFromList(graphs):
    unique_graphs = []
    for graph in graphs:
        hash = hashGraphContent(graph)
        if hash not in unique_graph_hashes:
            unique_graph_hashes.add(hash)
            unique_graphs.append(graph)
    return unique_graphs
```

```
def checkGraphsIdentical(g1, g2):
    # Check if graph number of nodes and number of edges is the same and if the node features are the same
    if (
        torch.equal(g1.ndata["feature"], g2.ndata["feature"])
        and torch.equal(g1.edges()[0], g2.edges()[0])
        and torch.equal(g1.edges()[1], g2.edges()[1])
    ):
        return True
    return False
```

```
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
graphs, labels = dgl.load_graphs(
    "../../Datasets/CIC-DDOS2019/My Preprocessing/Graphs_test/Size 20/0_UDP.dgl"
)
```

```
unique = True
unique_graphs = []
for i in range(len(graphs)):
    unique = True
    for j in range(0,i):
        if checkGraphsIdentical(graphs[i], graphs[j]):
            unique = False
            break
    if unique:
        unique_graphs.append(graphs[i])
print(f"Graphs: {len(graphs)}\nUnique: {len(unique_graphs)}")
```
